Single Variable DEM

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from PyEMD import EEMD

# 读取CSV文件
file_path = 'G:\\TOA_GC\\GCPost-Input\\GraphCast\\PostProcessing\\GraphCast_PostProcessing_Input_1-d.csv'  # 替换为你的CSV文件路径
df = pd.read_csv(file_path)


# 选择要分解的列
columns_to_decompose = ['spechumidity_500']

# 提取数据
data = df[columns_to_decompose].values

# 初始化EEMD对象参数
eemd = EEMD()
eemd.trials = 200  # EEMD的集合次数
eemd.noise_width = 0.5  # 噪声幅度

# 定义颜色列表
colors = ['r']

# 存储结果的DataFrame
result_df = pd.DataFrame()

# 对每列数据进行EEMD分解
for col_idx, col_name in enumerate(columns_to_decompose):
    signal = data[:, col_idx]
    
    # 调整信号长度为偶数
    if len(signal) % 2 != 0:
        signal = signal[:-1]
    
    # 执行EEMD分解
    imfs = eemd.eemd(signal)
    num_imfs = imfs.shape[0]
    
    # 添加IMFs到结果DataFrame
    for i in range(num_imfs):
        result_df[f"{col_name}_IMF_{i+1}"] = imfs[i]
    
    # 计算残差
    residual = signal - np.sum(imfs, axis=0)
    result_df[f"{col_name}_Residual"] = residual
    
    # 绘制结果
    t = np.arange(len(signal))  # 使用数据点索引作为时间轴
    plt.figure(figsize=(15, 12))
    plt.subplot(num_imfs + 2, 1, 1)
    plt.plot(t, signal, 'g')
    plt.title(f"Original Signal - {col_name}")

    for i in range(num_imfs):
        plt.subplot(num_imfs + 2, 1, i + 2)
        plt.plot(t, imfs[i], colors[i % len(colors)])
        plt.title(f"IMF {i+1} - {col_name}")

    plt.subplot(num_imfs + 2, 1, num_imfs + 2)
    plt.plot(t, residual, 'teal')
    plt.title(f"Residual - {col_name}")

    plt.tight_layout()
    plt.show()

# 保存结果到CSV文件
result_file_path = 'G:\\TOA_GC\\GCPost-Input\\VMD-GraphCast\\EEMD_results-station-Q500-1d.csv'
result_df.to_csv(result_file_path, index=False)
```
